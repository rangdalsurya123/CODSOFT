```python
import pandas as pd
import matplotlib.pyplot as plt

# Create sample sales data
data = {
    "Month": ["Jan", "Feb", "Mar", "Apr", "May", "Jun"],
    "Sales": [20000, 25000, 30000, 28000, 35000, 40000],
    "Profit": [4000, 5000, 7000, 6500, 8000, 9000]
}

df = pd.DataFrame(data)

# Display data
df
```

```python
# Plot Sales Trend
plt.figure()
plt.plot(df["Month"], df["Sales"])
plt.xlabel("Month")
plt.ylabel("Sales")
plt.title("Monthly Sales Trend")
plt.show()
```

```python
# Plot Profit Trend
plt.figure()
plt.plot(df["Month"], df["Profit"])
plt.xlabel("Month")
plt.ylabel("Profit")
plt.title("Monthly Profit Trend")
plt.show()
```
